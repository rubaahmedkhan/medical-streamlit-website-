# **Hangman**

```python
import random
from typing import List

# Function to generate a list of random words
def generate_random_words() -> List[str]:
    # Example larger pool of words
    all_words: List[str] = ['python', 'hangman', 'developer', 'computer', 'programming', 'algorithm', 'debugging', 'function', 'variable', 'loop', 'syntax', 'exception']
    random.shuffle(all_words)  # Shuffle the list to get a random order
    return all_words[:4]  # Get the first 4 words from the shuffled list

# Function to play Hangman
def hangman() -> None:
    words: List[str] = generate_random_words()  # Generate a random list of words for the game
    word: str = random.choice(words)  # Randomly choose a word from the list
    word_display: List[str] = ['_'] * len(word)  # Initial word display with blanks
    attempts: int = 6  # Number of attempts
    guessed_letters: List[str] = []  # List of guessed letters

    print("Welcome to Hangman!")

    while attempts > 0:
        print(f"\nWord: {' '.join(word_display)}")  # Show the word with blanks
        print(f"Guessed Letters: {', '.join(guessed_letters)}")
        print(f"Remaining Attempts: {attempts}")

        guess: str = input("Enter a letter: ").lower()  # Get user's guess
        if guess in guessed_letters:  # Check if letter was already guessed
            print("You already guessed that letter.")
            continue

        if len(guess) != 1 or not guess.isalpha():  # Validate input
            print("Invalid input! Please enter a single letter.")
            continue

        guessed_letters.append(guess)

        if guess in word:  # Correct guess
            print(f"Good job! '{guess}' is in the word.")
            for i in range(len(word)):
                if word[i] == guess:
                    word_display[i] = guess  # Update word display
        else:  # Incorrect guess
            attempts -= 1
            print(f"'{guess}' is not in the word.")

        if '_' not in word_display:  # Check if the word is fully guessed
            print(f"\nCongratulations! You've guessed the word: {word}")
            break

    if attempts == 0:  # If attempts are over
        print(f"\nGame Over! The word was: {word}")

# Start the game
hangman()
```
